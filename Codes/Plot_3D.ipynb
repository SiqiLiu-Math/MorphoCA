```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from ipywidgets import interactive, interact, fixed
import ipywidgets as widgets
import pickle

save_file_name = 'morpho3d2.pkl'
```

```python
with open(save_file_name,'rb') as pk:
    x_list,y_list,z_list,u_list,v_list,m3_list,m4_list = pickle.load(pk)
```

```python
def sphere(cx, cy, cz, r, resolution=20):
    u = np.linspace(0, 2*np.pi, resolution)
    v = np.linspace(0, np.pi, resolution)
    
    x = cx + r * np.outer(np.cos(u), np.sin(v))
    y = cy + r * np.outer(np.sin(u), np.sin(v))
    z = cz + r * np.outer(np.ones_like(u), np.cos(v))
    
    return x, y, z
```

```python
def plot3dsphere(t):
    xx = x_list[t]
    yy = y_list[t]
    zz = z_list[t]
    uu = u_list[t]
    vv = v_list[t]
    #vv = np.arange(len(xx))%4
    fig = go.Figure()
    for ii in range(len(xx)):
        x1,y1,z1 = sphere(xx[ii],yy[ii],zz[ii],uu[ii])
        fig.add_surface(
            x=x1, y=y1, z=z1,
            opacity=1,
            surfacecolor=np.full_like(x1,vv[ii]),
            coloraxis="coloraxis",
            showscale=False
        )
    fig.update_layout(
        coloraxis=dict(
            colorscale="Jet",
            colorbar=dict(title="Volume"),
            cmin=0,cmax=30
        )
    )

    fig.update_layout(
        scene = dict(
            xaxis = dict(
                showbackground=False,  # Remove background color behind x-axis
                showgrid=False,        # Remove x-axis gridlines
                zeroline=False,        # Remove x-axis zero line
                showticklabels=False, ticks="",range=[-8, 8]
            ),
            yaxis = dict(
                showbackground=False,  # Remove background color behind y-axis
                showgrid=False,        # Remove y-axis gridlines
                zeroline=False,        # Remove y-axis zero line
                showticklabels=False, ticks="",range=[-8, 8]
            ),
            zaxis = dict(
                showbackground=False,  # Remove background color behind z-axis
                showgrid=False,        # Remove z-axis gridlines
                zeroline=False,        # Remove z-axis zero line
                showticklabels=False, ticks="",range=[-8, 8]
            ),
            xaxis_title='',
            yaxis_title='',
            zaxis_title='',
            aspectmode="cube"
        ),
        scene_camera=dict(
            eye=dict(x=1.0, y=1.0, z=1.0)   # camera position
        ),
        width=600,height=500,
    )
    fig.show()
```

```python
interact(
    plot3dsphere,
    t=widgets.IntSlider(min=0,max=549,step=2,value=0),
);
```
